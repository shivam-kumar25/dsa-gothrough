```python
# GRPA 2


def findLargest (L):
    """
    This function finds the largest element in sorted and rotated list 'L' using binary search.

    Complexity :
    - Since we're reducing the search space by half in each step using a binary search, the time complexity is **O(log n)**.
    - The space complexity is **O(1)**, as no extra data structures are used.

    Approach :
    1. We use two pointers ('left_' and 'right_') to perform binary search.
    2. The key observation is:
       - If 'L[mid_] > L[right_]', then the pivot (largest element) is on the **right side**.
       - Otherwise, the pivot is on the **left side** or at 'mid_' itself.
    3. Once the search terminates, we return the **largest element**.

    """
    len_ = len(L)  # Get the length of the list.

    # Step 1: Initilize two pointers
    left_ , right_ = 0 , (len_-1)

    # Edge case: If the list contains only one element, that element is the largest.
    if len_ == 1:
        return L[0]  # Return the only element of the list.

    # Binary search loop to find the largest element in the rotated sorted list.
    while (left_ <= right_) :  # Continue searching untill left and right coverage.
        mid_ = (left_ + right_) // 2   # Compute the middle index of the current loop.

        # If mid_ is at the last index, set the next element to 0 (wrap around)
        # This ensures we correctly compare the fitst_ and last_ element in a rotated list.
        if mid_ == (len_-1) :
            next_to_mid_ = 0  # If mid is last index, next element is first index (circular check)
        else:
            next_to_mid_ = mid_ + 1  # Otherwise, the next element is simply (mid_+1)

        # If the middle element is grater than the next element, we have found the peak
        # Since the array is rotated, the largest element will always be grater than its next element.
        if L[mid_] > L[next_to_mid_]:
            return L[mid_]  # Return the largest element found

        # If the middle element is smaller than the first element of the list
        # This means the largest element must be in the left half of the list
        elif L[mid_] < L[0]:
            right_ = mid_ - 1  # Move the search range to the left half

        else:
            left_ = mid_ + 1  # Move the search range to left half


    # if the loop completes without returning, return the first element as a fallback
    # this happen in case, the list is not rotated at all
    return L[0]



print(findLargest([7, 8, 2, 4, 5, 6]))
```
