# Linear Search  (Na√Øve Search)

Linear search is a straightforward searching algorithm that checks each element in the list sequentially until the target value is found or the list is fully traversed.


```python

def linear_search(array_ , target_):
    for i in range(len(array_)):
        if array_[i] == target_ :
            return i

    return "Element Not Found"

list_1 = [3, 4, 1, 9, 56, 7, 9, 12, 13]
target_1 = 12
print(linear_search(list_1,target_1))




```
